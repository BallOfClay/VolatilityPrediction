```python
from datetime import datetime as dt
import numpy as np
import bs4 as Soup
import requests
import pandas as pd
```

```python
now = dt.now()
Year,WeekNum,DOW = now.isocalendar()
```

```python
print WeekNum
```

```python
histEconReports = []
for eWeek in np.arange(1,WeekNum+1):
    if eWeek < 10:
        eWeek = '0' + str(eWeek)
    histEconReports.append('http://biz.yahoo.com/c/ec/2014' + str(eWeek) + '.html') 
```

```python
res = requests.get(histEconReports[1])
```

```python
print histEconReports[1]
```

```python
soup = Soup.BeautifulSoup(res.content)
```

```python
import re
def cleanText(text):
#     print text
    c = ['-', 'NA']  
    text = re.sub('[%s]' % ''.join(c), '', text)
    if len(text) == 0:
        return ''
    # clean text...anystring that repeats 0 thru 9, return the first element
    cleaned = re.search("[0-9.]*",text).group(0)
    # if cleaned is empty string then try * to be sure
    if cleaned == '':
        cleaned = re.search("[0-9.]+",text).group(0)
    return cleaned
    
    
```

```python

import json
table = soup.findAll('table') #,{'class':'quoteContainer'})
briefing_forecast = ''
market_expects = ''
revised_from = ''
yahoo_data = []

trRows = table[5].findAll('tr')

for i, eRow in enumerate(trRows):
    listDF = []
    if i == 0:  # skip the header row
        continue
    cols = eRow.findAll('td')
    for eString in np.arange(0,3):
        listDF.append(cols[eString].string)
    # for columns 3 - 8 clean data
    for sIdx in np.arange(4,9):
        listDF.append(cleanText(cols[sIdx].string))
    yahoo_data.append(listDF)

df = pd.DataFrame(yahoo_data)
df.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://ethancheung@localhost:5432/test')
engine.connect()
table_name = 'stocknews_yahoocalendar'

df.to_sql(table_name, engine)

    
```
